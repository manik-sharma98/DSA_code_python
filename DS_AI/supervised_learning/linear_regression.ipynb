```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
pd_sal = pd.read_csv('Salary_Data.csv')
pd_sal.head()
```

```python
pd_sal.describe()
```

```python
plt.scatter(pd_sal['Salary'],pd_sal['YearsExperience'] ,color ='lightcoral')
plt.xlabel('Salary')
plt.ylabel('YOE')
plt.box(False)
```

```python
x = pd_sal.iloc[:,:1]
y = pd_sal.iloc[:,1:]
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
regersoor = LinearRegression()
regersoor.fit(X_train,y_train)
```

```python
y_pred_test = regersoor.predict(X_test)
y_pred_train = regersoor.predict(X_train)
```

```python
plt.plot(X_train,y_pred_train,color='red')
plt.scatter(X_train,y_train,color='black')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(False)
```
